```python
expense_list = []
monthly_budget = 0
total_monthly_expenses = 0
```

```python
def add_expense_details(date_entry, category_entry, amount_entry, description_entry):
  print("\nAdding new expense details.\n")
  expense_dict = {'date': date_entry, 'category': category_entry, 'amount': amount_entry, 'description': description_entry}
  expense_list.append(expense_dict)
  print("Expense added successfully.")
  return expense_dict
```

```python
add_expense_details("2025-12-01", "food", 100, "dinner")
```

```python
def view_expenses():
  print("\nViewing all expenses.\n")
  for expense in expense_list:
    if expense['date'] == '' or expense['category'] == '' or expense['amount'] == '' or expense['description'] == '':
      #skip the entry
      print("There is an incomplete entry for:", expense)
    else:
      print(expense)
```

```python
view_expenses()
```

```python
add_expense_details("2025-12-02", "taxi", 50, '')
```

```python
view_expenses()
```

```python
def set_monthly_budget(m_budget):
  global monthly_budget
  monthly_budget = m_budget
  print("Monthly budget set to:", monthly_budget)
```

```python
print(monthly_budget)
set_monthly_budget(400)
print(monthly_budget)
```

```python
def calculate_total_monthly_expenses():
  print("Calculating total monthly expenses.")
  global total_monthly_expenses
  total_monthly_expenses = 0
  for expense in expense_list:
    total_monthly_expenses += int(expense['amount'])
  remaining_balance = monthly_budget - total_monthly_expenses
  print("Total monthly expense is:", total_monthly_expenses, "monthly budget is:", monthly_budget, "\n")
  if remaining_balance < 0:
    print("Warning: you have exceeded your budget!")
  else:
    print("Remaining balance:", remaining_balance)
```

```python
calculate_total_monthly_expenses()
```

```python
def save_expenses():
  print("\nSaving expenses to file.\n")
  with open("expenses.txt", "w") as f:
    for expense in expense_list:
      f.write(f"{expense['date']},{expense['category']},{expense['amount']},{expense['description']}\n")
    f.close()
    print("Expenses saved successfully.")
```

```python
save_expenses()
```

```python
def load_expenses():
  print("\nLoading expenses from file.\n")
  with open("expenses.txt", "r") as f:
    for line in f:
      new_date, new_category, new_amount, new_description = line.strip().split(',')
      expense_dict = {'date': new_date, 'category': new_category, 'amount': int(new_amount), 'description': new_description}
      expense_list.append(expense_dict)
  f.close()
  print("Expenses loaded successfully.")
```

```python
load_expenses()
```

```python
print(expense_list)
```

```python
def main():
  while True:
    print("---------------------------------------------------")
    print("Please choose an option by entering a number")
    print("1. Add an expense")
    print("2. View expenses")
    print("3. Track the budget")
    print("4. Save expenses to a file")
    print("5. Save expenses and exit")
    print("6. Load expenses")
    choice = input("Enter your choice: ")

    if choice == "1":
      date_entry = input("Enter the date of your expense in YYYY-MM-DD format: ")
      category_entry = input("Enter the category of your expense: ")
      amount_entry = input("Enter the amount of your expense: ")
      description_entry = input("Enter a description of your expense: ")
      add_expense_details(date_entry, category_entry, amount_entry, description_entry)
    elif choice == "2":
      view_expenses()
    elif choice == "3":
      m_budget = input("Enter your monthly budget: ")
      set_monthly_budget(int(m_budget))
      calculate_total_monthly_expenses()
    elif choice == "4":
      save_expenses()
    elif choice == "5":
      save_expenses()
      print("------------------------EXITING--------------------")
      break
    elif choice == "6":
      load_expenses()
    else:
      print("Invalid choice. Please try again.")

if __name__ == "__main__":
  main()
```
